## **1. Generate Two Vectors with 15 Random Floats and Plot them on the same graph**


```python
import numpy as np
import matplotlib.pyplot as mat

# Creating two vectors of 15 random floats each
x = np.arange(15) 
y1 = np.random.rand(15)  
y2 = np.random.rand(15)  

# Plot them on the same graph
mat.plot(x, y1, marker='o', linestyle='-', color='b', label='Vector 1')
mat.plot(x, y2, marker='s', linestyle='--', color='r', label='Vector 2')

# Adding labels and title
mat.xlabel("Index")
mat.ylabel("Value")
mat.title("Plot of Two Random Vectors")
mat.legend()
mat.grid(True)

mat.show()
```

## Discussion of Effects

### Matrix Addition (A + B)
- Each number in **Matrix A** is added to the matching number in **Matrix B**.

- The **bar chart** shows how the numbers increase.

### Matrix Subtraction (A - B)
- Each number in **Matrix B** is subtracted from the corresponding number in **Matrix A**.
- If **B** is larger than **A**, the result is negative.
- The **bar chart** shows decreases and negatives.

### Matrix Multiplication (A * B)
- This is **not element-wise** multiplication but follows **matrix rules**.
- Rows of **A** multiply with columns of **B**, creating a **new matrix**.
- The **bar chart** shows major value changes.
