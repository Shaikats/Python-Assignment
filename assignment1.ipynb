### **1. Generate Two Vectors with 15 Random Floats**


```python
import numpy as np
import matplotlib.pyplot as mat

# Generate two vectors of 15 random floats each
x = np.arange(15) 
y1 = np.random.rand(15)  
y2 = np.random.rand(15)  

# Plot them on the same graph
mat.plot(x, y1, marker='o', linestyle='-', color='b', label='Vector 1')
mat.plot(x, y2, marker='s', linestyle='--', color='r', label='Vector 2')


mat.xlabel("Index")
mat.ylabel("Value")
mat.title("Plot of Two Random Vectors")
mat.legend()
mat.grid(True)

mat.show()
```
